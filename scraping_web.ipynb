```python
import requests
from bs4 import BeautifulSoup

url = 'https://exampleblog.com/post'
url = 'https://community.verizon.com/t5/Verizon-Community-Blog/bg-p/mobile-home-blogs'
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue'
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'

page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
# print(soup)
```

```python
t1 = soup.find('table', class_ = 'wikitable sortable')
# t1
```

```python
table = soup.find_all('table')[1]
```

```python
col_names_all = table.find_all('th')
col_names = [c.text.strip() for c in col_names_all]
print(col_names)
```

```python
import pandas as pd
df = pd.DataFrame(columns=col_names)
df
```

```python
col_data = table.find_all('tr')
for row in col_data[1:]:
  row_data = row.find_all('td')
  individual_row_data = [ data.text.strip() for data in row_data ]
  # print(individual_row_data)

  df.loc[len(df)] = individual_row_data
```

```python
df
```

```python
df.to_csv('companies.csv', index=False)
```
